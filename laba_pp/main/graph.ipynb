```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('C:/PYTHON/laba_pp/file/result.txt', sep=' ', header=None)
data = data.drop([0, 2, 4], axis=1)
data.columns = ['Size', '№', 'Time']
```

```python
data.head(3)
```

```python
sns.scatterplot(x='Size', y='Time', hue='№', data=data)

plt.xlabel('Size')
plt.ylabel('Time')
plt.title('Зависимость времени от размера для каждого значения №')

plt.show()
```

```python
sns.pointplot(x='Size', y='Time', hue='№', data=data)

plt.xlabel('Size')
plt.ylabel('Time')
plt.title('Зависимость времени от размера для каждого значения №')
plt.savefig('C:/PYTHON/laba_pp/file/plot.png')
plt.show()
```
