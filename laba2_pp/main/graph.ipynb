```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df1 = pd.DataFrame(columns=['Size', '№', 'Time'])

with open('file\\result_2.txt', 'r') as file:
    for line in file:
        data = line.split()
        size = int(data[1])
        num = int(data[3])
        time = float(data[5])

        df1 = df1.append({'Size': size, '№': num, 'Time': time}, ignore_index=True)
```

```python
df3 = pd.DataFrame(columns=['Size', '№', 'Time'])

with open('file\\result_8.txt', 'r') as file:
    for line in file:
        data = line.split()
        size = int(data[1])
        num = int(data[3])
        time = float(data[5])

        df3 = df3.append({'Size': size, '№': num, 'Time': time}, ignore_index=True)
```

```python
df4 = pd.DataFrame(columns=['Size', '№', 'Time'])

with open('file\\result_12.txt', 'r') as file:
    for line in file:
        data = line.split()
        size = int(data[1])
        num = int(data[3])
        time = float(data[5])

        df4 = df4.append({'Size': size, '№': num, 'Time': time}, ignore_index=True)
```

```python
fig, ax = plt.subplots()

ax.plot(df1['Size'], df1['Time'], marker='o', linestyle='-', label='2 потока')
ax.plot(df2['Size'], df2['Time'], marker='o', linestyle='-', label='4 потока')
ax.plot(df3['Size'], df3['Time'], marker='o', linestyle='-', label='8 потоков')
ax.plot(df4['Size'], df4['Time'], marker='o', linestyle='-', label='12 потоков')

ax.set_xlabel('Size')
ax.set_ylabel('Time')
ax.set_title('График времени от размера')
ax.legend(title='Датафреймы')
plt.grid(True)
plt.show()
```
