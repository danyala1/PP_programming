```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.DataFrame(columns=['Size', '№', 'Time'])

with open('file//result.txt', 'r') as file:
    for line in file:
        data = line.split()
        size = int(data[1])
        num = int(data[3])
        time = float(data[5])

        df = df.append({'Size': size, '№': num, 'Time': time}, ignore_index=True)
```

```python
df.head(5)
```

```python
df.groupby('Size')['Time'].describe()
```

```python
sns.scatterplot(x='Size', y='Time', hue='№', data=df)

plt.xlabel('Size')
plt.ylabel('Time')
plt.title('Зависимость времени от размера для каждого значения №')

plt.show()
```

```python
df2 = pd.DataFrame(columns=['Size', '№', 'Time'])

with open('file//result_old.txt', 'r') as file:
    for line in file:
        data = line.split()
        size = int(data[1])
        num = int(data[3])
        time = float(data[5])

        df2 = df2.append({'Size': size, '№': num, 'Time': time}, ignore_index=True)
```
